```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import pandas as pd
import numpy as np
```

```python
# Get the binned and summed occupancy data
bathroom_occupancy = pd.read_csv('binned_and_summed_occupancy/bathroom_occupancy_binned_and_summed.csv')
bedroom_child_occupancy = pd.read_csv('binned_and_summed_occupancy/bedroom_child_occupancy_binned_and_summed.csv')
bedroom_parents_occupancy = pd.read_csv('binned_and_summed_occupancy/bedroom_parents_occupancy_binned_and_summed.csv')
hallway_occupancy = pd.read_csv('binned_and_summed_occupancy/hallway_occupancy_binned_and_summed.csv')
office_occupancy = pd.read_csv('binned_and_summed_occupancy/office_occupancy_binned_and_summed.csv')
```

```python
# Plot the bathroom occupancy
x = bathroom_occupancy['time']
y = bathroom_occupancy['total']

fig = plt.figure(figsize =(21, 9))
plt.xticks(rotation=90)

plt.bar(x,y) # A bar chart
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.show()
```

```python
# Plot the child bedroom occupancy
x = bedroom_child_occupancy['time']
y = bedroom_child_occupancy['total']

fig = plt.figure(figsize =(21, 9))
plt.xticks(rotation=90)

plt.bar(x,y,align='center') # A bar chart
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.show()
```
